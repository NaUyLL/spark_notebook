```scala
import org.apache.spark.{SparkConf, SparkContext}
import org.apache.spark.sql.SparkSession
import org.apache.spark.sql.functions.lit
```

```scala
val conf=new SparkConf()
      .setMaster("local")//启动本地化计算
      .setAppName("testRdd")//设置本程序名称
val sc=new SparkContext(conf)
val spark = SparkSession.builder().appName("json_app").getOrCreate()
```

```scala
val csv1=spark.read.option("header", "true").csv("../data/input.csv")//读取本地文件
```

```scala
val col = csv1.columns
```

```scala
var index = col.indexWhere(_=="Q1.9_1")
println(index)
```

```scala
var index = col.indexWhere(_=="activities_1")
println(index)
```

```scala
val Ques = col.slice(col.indexWhere(_=="activities_1"), col.length)
println(Ques.length
```

```scala
var newDF = csv1.select("ResponseId","EndDate",Ques(0)).toDF("ResponseId","EndDate", "Value").withColumn("Question", lit(Ques(0)))
```

```scala
for(i<-(1 until Ques.length)){
    var seDF = csv1.select("ResponseId","EndDate",Ques(i)).toDF("ResponseId","EndDate", "Value").withColumn("Question", lit(Ques(i)))
    newDF = newDF.union(seDF)
}
```
